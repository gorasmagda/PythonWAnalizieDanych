```python
#zadanie 11 - oblicza n-t liczb Fibonacciego (n du偶e).

def zad11(x):
    a = 0 
    b = 1
    if x == 1:
        return a
    elif x == 2:
        return b
    else: 
        for i in range(3,x+1):
            c = a+b 
            a = b 
            b = c
        return b 
zad11(15)

#Funkcja liczy n-t liczb Fibonacciego (n mae)

def fib(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print(fib(10))
```

```python
#zadanie 19 - dla zadanego string zwr贸ci jego posta w notacji wielbdziej, np. dla "ala ma pythona" otrzymamy 'alaMaPythona'.

def zad19(x):
    words = x.split()
    words = [words[0].lower()] + [word.capitalize() for word in words[1:]]
    return "".join(words)
zad19("ala ma pythona")
```

```python
#zadanie 20 - dla string w postaci wielbdziej zwr贸ci jego posta "normaln", np. dla 'alaMaPythona' otrzymamy "ala ma pythona".

def zad20(x):
    to_split = [i for i in x if i.isupper()]
    letters = list(x)
    normal_word = letters[0].lower()
    for i in letters[1:]:
        if i in to_split:
            normal_word += " "
            normal_word += i.lower()
        else:
            normal_word += i
    return normal_word
zad20("alaMaKotaAKotMaAle")
```

```python
#zadanie 21 - zgodnie z metod babilosk obliczy przybli偶enie pierwiastka kwadratowego danej liczby.

def zad21(x):
    g = x/2.0
    g2 = g + 1
    while(g != g2):
        n = x/ g
        g2 = g
        g = (g + n)/2

    return g
zad21(5)
```

```python
#zadanie 22 - przybli偶a warto   (metoda Monte-Carlo).
import random as r

def zad22():
    circle_points = 0
    square_points = 0
    interval = 1000
    for i in range(interval*2):
        rand_x = r.uniform(-1,1)
        rand_y = r.uniform(-1,1)
        origin_dist = rand_x**2 + rand_y**2
        if origin_dist <=1:
            circle_points+= 1
        square_points +=1
        pi = 4*circle_points/square_points
    return pi
zad22()
```
